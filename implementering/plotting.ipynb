```python
import matplotlib.pyplot as plt
```

```python
x1 = [0, 1, 2, 3, 4, 5]
y1 = []
for tall in x1:
    y1.append(3*tall + 2)
plt.plot(x1,y1) # plot gjør at det blir en linjeplt.scatter(x1,y1) # scatter gjør at det blir punkterplt.show()
plt.scatter(x1,y1) # scatter gjør at det blir punkterplt.show()
plt.show()
```
