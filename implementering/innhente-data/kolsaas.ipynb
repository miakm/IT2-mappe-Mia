```python
import matplotlib.pyplot as plt
```

```python
fil = open("kolsaas.csv")
data = fil.read()
```

```python
linjer = data.split("\n")
linjer
tur = []
for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)
tur[0][1]



```

```python
x = []
y = []
lengde = []
bredde = []

for i in tur[1:]:
    lengde.append(float(i[2]))
    bredde.append(float(i[1]))
    altitude = float(i[3])
    time = float(i[0])
    x.append(time)
    y.append(altitude)

plt.scatter(x,y)
plt.style.use("fivethirtyeight")
plt.show()
```
