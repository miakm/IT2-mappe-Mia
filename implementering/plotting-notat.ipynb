```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 3*x**2

x = []
y = []

for i in range(20):
    x.append(i)
    y.append(f(i))


plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,200,10000)
y = f(x)

plt.plot(x,y)
plt.style.use("fivethirtyeight")
plt.title("$f(x)=3x^2$")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```
